```python
import pandas as pd

# Load dataset with error handling
try:
    df = pd.read_csv("iris.csv")  # replace with your dataset
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: Dataset file not found.")

# Display the first few rows
print(df.head())

# Check data types and missing values
print(df.info())
print(df.isnull().sum())

# Clean the dataset by dropping or filling missing values
df = df.dropna()  # or use df.fillna(value)

# Summary statistics
print(df.describe())

# Grouping by a categorical column and finding mean
grouped = df.groupby("species").mean()  # adjust column to your dataset
print(grouped)

import matplotlib.pyplot as plt
import seaborn as sns

# Line Chart (fake example if you have a time column)
# df.plot(x='date', y='sales', kind='line')

# Bar Chart
sns.barplot(data=df, x='species', y='petal_length')  # adjust to your columns
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()

# Histogram
df['sepal_length'].hist(bins=20)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot
plt.scatter(df['sepal_length'], df['petal_length'])
plt.title('Sepal vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()

```
